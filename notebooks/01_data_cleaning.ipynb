#### Fill null value for 'conditions' feature:


```python
# Fill null value for 'conditions' feature as 'Unknown'

clinical_trial_dataframe['conditions'] = clinical_trial_dataframe['conditions'].fillna("Unknown")
print("Sum of missing values:")
print(clinical_trial_dataframe.isnull().sum())
print("-"*50)
```

#### Fill null values for 'sex' feature:

```python
# Fill null values in 'sex' feature using groupby option with 'source_condition_query' and 'sex' using mode function:
print(clinical_trial_dataframe['sex'].value_counts())
print("-"*50)
clinical_trial_dataframe['sex'] = clinical_trial_dataframe['sex'].fillna(
                                    clinical_trial_dataframe.groupby('source_condition_query')['sex'].
                                    transform(lambda x: x.mode()[0]))

# To find null values in each count:
print("\nSum of misisng values: ")
print(clinical_trial_dataframe.isnull().sum())
print("-" * 50)
print(clinical_trial_dataframe['sex'].unique())
print("-" * 50)
print(clinical_trial_dataframe['sex'].value_counts())

```

#### fill null values in 'phase'

```python
# Filling null values in 'phase' feature:
print(clinical_trial_dataframe['phase'].value_counts())
print("-" * 50)

# clinical_trial_dataframe['phase'] = clinical_trial_dataframe['phase'].fillna("NA")
clinical_trial_dataframe['phase'] = clinical_trial_dataframe['phase'].fillna("Not_Applicable")


# To find null values in each count:
print("\nSum of misisng values: ")
print(clinical_trial_dataframe.isnull().sum())
print("-" * 50)
print(clinical_trial_dataframe['phase'].unique())
print("-" * 50)
print(clinical_trial_dataframe['phase'].value_counts())


```

#### Fill null value in 'healthy_volunteer'

```python
# Filling null values in 'healthy_volunteers' feature:
print(clinical_trial_dataframe['healthy_volunteers'].value_counts())
print("-" * 50)

clinical_trial_dataframe['healthy_volunteers'] = clinical_trial_dataframe['healthy_volunteers'].fillna(
                        clinical_trial_dataframe.groupby(['study_type', 'phase'])['healthy_volunteers']
                        .transform(lambda x: x.mode()[0]))


# To find null values in each count:
print("\nSum of misisng values: ")
print(clinical_trial_dataframe.isnull().sum())
print("-" * 50)
print(clinical_trial_dataframe['healthy_volunteers'].unique())
print("-" * 50)
print(clinical_trial_dataframe['healthy_volunteers'].value_counts())
```

#### Fill null values in 'eligibility_criteria'

```python
# Fill null values in 'eligibility_criteria' as 'No_Criteria_Mentioned':

clinical_trial_dataframe['eligibility_criteria'] = clinical_trial_dataframe['eligibility_criteria'].fillna("No_Criteria_Mentioned")

# To find null values in each count:
print("\nSum of misisng values: ")
print(clinical_trial_dataframe.isnull().sum())
print("-" * 50)

print("\nNo of Unique values in eligibility_criteria:", clinical_trial_dataframe['eligibility_criteria'].nunique())
print("-" * 50)
```

#### Fill null values in 'Official_title'

```python
# Fill null values in 'official_title' by filling same value as it is in 'title':
clinical_trial_dataframe['official_title'] = clinical_trial_dataframe['official_title'].fillna(clinical_trial_dataframe['title'])

# To find null values in each count:
print("\nSum of misisng values: ")
print(clinical_trial_dataframe.isnull().sum())
print("-" * 50)

print("\nNo of Unique values in official_title:", clinical_trial_dataframe['official_title'].nunique())
print("-" * 50)
```

#### Fill null values for 'interventions':

```python
# Fill null values in 'interventions' as 'No_Interventions_Mentioned':
clinical_trial_dataframe['interventions'] = clinical_trial_dataframe['interventions'].fillna("No_Interventions_Mentioned")

# To find null values in each count:
print("\nSum of misisng values: ")
print(clinical_trial_dataframe.isnull().sum())
print("-" * 50)

print("\nNo of Unique values in interventions:", clinical_trial_dataframe['interventions'].nunique())
print("-" * 50)

```

#### Fill null values for 'minimum_age' and 'maximum_age'

```python
print(clinical_trial_dataframe['minimum_age'].str.extract(r'(\D+)$')[0].unique())
print(clinical_trial_dataframe['maximum_age'].str.extract(r'(\D+)$')[0].unique())
```

#### split minimum age and maximum age features separatly like numbers and units separately


```python
# spliting the minimum age into min_age and min_age_units:
# st.split(" ", n=1, expand=True), split is done based on space, no of split is n = 1.

# Minimum_age:
clinical_trial_dataframe[['min_age' , 'min_age_unit']] = (clinical_trial_dataframe['minimum_age']
                                                            .str.split(" ", n =1, expand=True))

# Maximum age:
clinical_trial_dataframe[["max_age", "max_age_unit"]] = (clinical_trial_dataframe["maximum_age"]
                                                            .str.split(" ", n=1, expand=True))

# Changing datatype for min_age and max_age as int  
clinical_trial_dataframe['min_age'] = clinical_trial_dataframe['min_age'].astype(float)
clinical_trial_dataframe['max_age'] = clinical_trial_dataframe['max_age'].astype(float)


# To find null values in each count:
print("\nSum of misisng values: ")
print(clinical_trial_dataframe.isnull().sum())
print("-" * 50)


# Normalising the strings in min_unit and max_unit like from 'years' to 'year', 'months' to 'month'
# str.strip()  - used to remove unwanted spaces before and after each string.
# str.lower()  - used to change entire string in lowercase.
# str.rstrip('s') - used to remove letter from right side for each string.
# str.lstrip('s') - used to remove letter from left side for each string.
clinical_trial_dataframe['min_age_unit'] = clinical_trial_dataframe['min_age_unit'].str.strip().str.lower().str.rstrip('s')
clinical_trial_dataframe['max_age_unit'] = clinical_trial_dataframe['max_age_unit'].str.strip().str.lower().str.rstrip('s')


# Conversion:
conversion = {'year': 1, 
              'month': 12, 
              'week': 52, 
              'day': 365, 
              'hour': 8760, 
              'minute': 525600}

clinical_trial_dataframe['minimum_age_years'] = (clinical_trial_dataframe['min_age'] / clinical_trial_dataframe['min_age_unit'].map(conversion)).round()
clinical_trial_dataframe['maximum_age_years'] = (clinical_trial_dataframe['max_age'] / clinical_trial_dataframe['max_age_unit'].map(conversion)).round()

#Filling null values for 'minimum_age_years' as '0' & 'maximum_age_years' as '120' as human life span of living normally which highest:
clinical_trial_dataframe['minimum_age_years'] = clinical_trial_dataframe['minimum_age_years'].fillna(0)
clinical_trial_dataframe['maximum_age_years'] = clinical_trial_dataframe['maximum_age_years'].fillna(120)

# To find null values in each count:
print("\nSum of misisng values: ")
print(clinical_trial_dataframe.isnull().sum())
print("-" * 50)

```

#### Removing unwanted Features:

```python
# To find null values in each count:
print("\nSum of misisng values before droping unwanted features: ")
print(clinical_trial_dataframe.isnull().sum())
print("-" * 50)

# Remove unwanted columns:
clinical_trial_dataframe = clinical_trial_dataframe.drop(
                        ["minimum_age","maximum_age","min_age","min_age_unit","max_age","max_age_unit"], axis = 1)


# To find null values in each count:
print("\nSum of misisng values after droping unwanted features: ")
print(clinical_trial_dataframe.isnull().sum())
print("-" * 50)



```
